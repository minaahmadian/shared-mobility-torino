```python
import pandas as pd
import os

# مسیر فولدر (حتماً مسیر واقعی خودت رو بذار)
folder_path = "/Users/mina/Desktop/Transport/OPERATORE_A_Lime/Torino_Corse24-25_MENSILI_senza_percorso"

# گرفتن فقط فایل‌های csv
csv_files = sorted([
    f for f in os.listdir(folder_path)
    if f.endswith(".csv")
])

print(f"{len(csv_files)} CSV files found")

# خوندن و ادغام
dfs = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    df["source_file"] = file   # خیلی مهم، نگهش دار
    dfs.append(df)

df_16merged_lime = pd.concat(dfs, ignore_index=True)

df_16merged_lime .head()
```

```python
import folium
from folium.plugins import HeatMap
import pandas as pd
import json
```

```python
center_lat = df_all_lime['LATITUDINE_INIZIO_CORSA'].mean()
center_lon = df_all_lime['LONGITUTIDE_INIZIO_CORSA'].mean()
```

```python
m_start = folium.Map(location=[center_lat, center_lon], zoom_start=12)

start_points = df_all_lime[
    ['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']
].dropna().values.tolist()

HeatMap(start_points, radius=8).add_to(m_start)

m_start.save("lime_start_heatmap.html")
```

```python
m_end = folium.Map(location=[center_lat, center_lon], zoom_start=12)

end_points = df_all_lime[
    ['LATITUDINE_FINE_CORSA', 'LONGITUTIDE_FINE_CORSA']
].dropna().values.tolist()

HeatMap(end_points, radius=8).add_to(m_end)

m_end.save("lime_end_heatmap.html")
```

```python
df_all_lime['HOUR'] = df_all_lime['DATAORA_INIZIO'].dt.hour

day_df = df_all_lime[(df_all_lime['HOUR'] >= 7) & (df_all_lime['HOUR'] < 19)]
night_df = df_all_lime[(df_all_lime['HOUR'] < 7) | (df_all_lime['HOUR'] >= 19)]
```

```python
m_day = folium.Map(location=[center_lat, center_lon], zoom_start=12)

HeatMap(
    day_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8
).add_to(m_day)

m_day.save("lime_day_heatmap.html")
```

```python
m_night = folium.Map(location=[center_lat, center_lon], zoom_start=12)

HeatMap(
    night_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8
).add_to(m_night)

m_night.save("lime_night_heatmap.html")
```

```python
df_all_lime['WEEKDAY'] = df_all_lime['DATAORA_INIZIO'].dt.weekday
weekday_df = df_all_lime[df_all_lime['WEEKDAY'] < 5]
weekend_df = df_all_lime[df_all_lime['WEEKDAY'] >= 5]
```

```python
m_weekday = folium.Map(location=[center_lat, center_lon], zoom_start=12)

HeatMap(
    weekday_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8
).add_to(m_weekday)

m_weekday.save("lime_weekday_heatmap.html")
```

```python
m_weekend = folium.Map(location=[center_lat, center_lon], zoom_start=12)

HeatMap(
    weekend_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8
).add_to(m_weekend)

m_weekend.save("lime_weekend_heatmap.html")
```

```python
sample_routes = df_all_lime.sample(5000, random_state=42)
```

```python
m_routes = folium.Map(location=[center_lat, center_lon], zoom_start=12)

for _, row in sample_routes.iterrows():
    try:
        geo = json.loads(row['PERCORSO'])
        coords = geo['coordinates']
        coords_latlon = [(c[1], c[0]) for c in coords]

        folium.PolyLine(
            coords_latlon,
            color='blue',
            weight=1,
            opacity=0.3
        ).add_to(m_routes)
    except:
        continue

m_routes.save("lime_route_intensity.html")
```

```python
top_origins = (
    df_all_lime
    .groupby(['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA'])
    .size()
    .reset_index(name='count')
    .sort_values('count', ascending=False)
    .head(30)
)
```

```python
m_top = folium.Map(location=[center_lat, center_lon], zoom_start=12)

for _, row in top_origins.iterrows():
    folium.CircleMarker(
        location=[row['LATITUDINE_INIZIO_CORSA'], row['LONGITUTIDE_INIZIO_CORSA']],
        radius=row['count'] / 200,
        popup=f"Trips: {row['count']}",
        fill=True
    ).add_to(m_top)

m_top.save("lime_top_origins.html")
```

```python
m_multi = folium.Map(location=[center_lat, center_lon], zoom_start=12)

HeatMap(
    weekday_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8,
    name='Weekday'
).add_to(m_multi)

HeatMap(
    weekend_df[['LATITUDINE_INIZIO_CORSA', 'LONGITUTIDE_INIZIO_CORSA']]
    .dropna().values.tolist(),
    radius=8,
    name='Weekend'
).add_to(m_multi)

folium.LayerControl().add_to(m_multi)

m_multi.save("lime_weekday_weekend_map.html")
```
