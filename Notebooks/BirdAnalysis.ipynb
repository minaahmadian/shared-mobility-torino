```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # radius of Earth in km
    
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    return R * c
```

```python
df_2024_bird_clean['HAV_DIST_KM'] = df_2024_bird_clean.apply(
    lambda row: haversine(row['LATITUDINE_INIZIO_CORSA'],
                          row['LONGITUTIDE_INIZIO_CORSA'],
                          row['LATITUDINE_FINE_CORSA'],
                          row['LONGITUTIDE_FINE_CORSA']),
    axis=1
)
```

```python
df_2025_bird_clean['HAV_DIST_KM'] = df_2025_bird_clean.apply(
    lambda row: haversine(row['LATITUDINE_INIZIO_CORSA'],
                          row['LONGITUTIDE_INIZIO_CORSA'],
                          row['LATITUDINE_FINE_CORSA'],
                          row['LONGITUTIDE_FINE_CORSA']),
    axis=1
)
```

```python
df_2024_final['DATAORA_INIZIO'] = pd.to_datetime(df_2024_final['DATAORA_INIZIO'])
df_2025_final['DATAORA_INIZIO'] = pd.to_datetime(df_2025_final['DATAORA_INIZIO'])
```

```python
weekly_2024 = df_2024_final.resample(
    'W', on='DATAORA_INIZIO'
).size()

weekly_2025 = df_2025_final.resample(
    'W', on='DATAORA_INIZIO'
).size()
```

```python
plt.figure(figsize=(14,5))
plt.plot(weekly_2024, label='2024', alpha=0.8)
plt.plot(weekly_2025, label='2025', alpha=0.8)

plt.title('Weekly Mobility Trends')
plt.xlabel('Week')
plt.ylabel('Number of Trips')
plt.legend()
plt.grid(True)
plt.show()
```

```python
monthly_2024 = df_2024_final.resample(
    'ME', on='DATAORA_INIZIO'
).size()

monthly_2025 = df_2025_final.resample(
    'ME', on='DATAORA_INIZIO'
).size()
```
